```python
import geopandas

locations = geopandas.tools.geocode('Moscow, Russia', provider='Yandex')
loc=locations.geometry[0]
[loc.y, loc.x]
```

```python
import folium

folium_map = folium.Map(location=[loc.y, loc.x])
marker = folium.Marker(location=[loc.y, loc.x], popup='Marker 1')
marker.add_to(folium_map)
folium_map
```
